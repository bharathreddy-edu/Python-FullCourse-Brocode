```python

# List  --->  ORDERED and CHANGABLE

vegetables = ["asparagus" ,"okra", "sorakaya", "dondakaya", "carrot", "chamagadda"]
print(vegetables)
#how to access element at a position, index start from positon at 0
print(vegetables[3])

#what happens if i give index as 10 for above list
#print(vegetables[10])
# this give error

print("printing vegetables via loop")

for vegies in vegetables:
    print(vegies, end="--------->")
    
print("-------------------------------------")  
print("-------------------------------------")  
vegetables.append("gongura")
print(vegetables)

vegetables.remove("sorakaya")
print(vegetables)

vegetables.append("cabbage")
print(vegetables)

vegetables.insert(0,"sukkakura")
print(vegetables)





print("----------------444444444444444444------------------------")
if "cabbage" in vegetables:
    print("cabbage is in the list")

print("----------------444444444444444444------------------------")



vegetables.pop()
print(vegetables)

print("----------------444444444444444444------------------------")
if "cabbage" not in vegetables:
    print("Cabbage is not there, pop() removed it")

print("----------------444444444444444444------------------------")


vegetables.sort()
print(vegetables)



# Duplicates are allowed
vegetables.append("asparagus")
print(vegetables)







vegetables.clear()
print(vegetables)

```

```python
# 2/3/4/....n dimensional list and print it
#find what matrix can be formed ? and print it

starters = ["chicke-lolipop", "chicken65", "califlowerfry", "fishfry", "pannerTikka", "sweetcorn"]
main_course = ["biryani", "curry_naan", "friedRice", "kottuparatha"]
desserts = ["gulabjamun", "custard", "falooda", "rasmalai", "pudding"]


menucombo = [starters, main_course, desserts]

templist = []
for i in range(len(menucombo)):
    templist.append(len(menucombo[i]))
templist.sort(reverse=True)
print("--------------------- Matrix Type----------------------")
print("what kind of matrix this gonna be")
print(f"Row\tX\tColumn\n{len(menucombo)}\tX\t{templist[0]}")
print("--------------------- Matrix Print----------------------")
for i in menucombo:
    print(f"{i}",end="\n")
print("--------------------- Matrix Print----------------------")
```
